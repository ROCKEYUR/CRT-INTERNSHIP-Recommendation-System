```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
movies.shape
```

```python
credits.head(1)
```

```python
credits.shape
```

```python
movies = movies.merge(credits,on='title')
```

```python
#Since 93.89% movies in this dataset have English as original language, we won't consider this
movies['original_language'].value_counts()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[0:10]
```

```python
import difflib

def recommend(movie):
    matches = difflib.get_close_matches(movie, new_df['title'], n=1, cutoff=0.5)
    if matches:
        matched_movie = matches[0]
        movie_index = new_df[new_df['title'] == matched_movie].index[0]
        distances = sorted(list(enumerate(similarity[movie_index])), reverse=True, key=lambda x: x[1])
        for i in distances[0:10]:
            print(new_df.iloc[i[0]].title)
    else:
        print("No matching movie found.")
```

```python
while True:
    user_input = input("Enter a movie name: ")
    if user_input.lower() == 'end':
        break
    recommend(user_input)

print("Thank you for using the recommendation system!")
```
